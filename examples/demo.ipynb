# Hubeau Data Client - General Demo

This notebook demonstrates how to use the unified `hubeau-data` client library to query French water APIs.


```python
from hubeau_data.client import HubeauClient
from hubeau_data.models.qualite_rivieres import StationPcParams
from hubeau_data.models.hydrometrie import ObservationTrParams

# Initialize the unified client
client = HubeauClient()
```

## 1. 'Qualité Rivières' API (Water Quality)

We retrieve the first 3 water quality measurement stations located in the Paris area (Department 75).

```python
# Query the Qualité Rivières API
# Fetching stations in Paris (75) area
stations_qr = client.qualite_rivieres.get_stations(
    params=StationPcParams(code_departement=["75"], size=3)
)
print(f"✓ Found {len(stations_qr)} stations in Paris area.\n")
for i, station in enumerate(stations_qr, 1):
    print(f"Station #{i}: {station.code_station} - {station.libelle_station}")
    print(f"  Location: {station.libelle_commune} ({station.code_commune})")
    print(f"  URI: {station.uri_station}\n")
```

## 2. 'Hydrométrie' API (Water Levels and Flows)

We retrieve the latest real-time observations for a specific reference station (station `Y120201001`).

```python
target_station = "Y120201001"

# Query the Hydrométrie API
obs_params = ObservationTrParams(code_entite=[target_station], size=3)
observations = client.hydrometrie.get_observations_tr(params=obs_params)


print(f"✓ Retrieved {len(observations)} real-time observations for station {target_station}.\n")
for i, obs in enumerate(observations, 1):
    print(f"Obs #{i} | Date: {obs.date_obs} | Value: {obs.resultat_obs} {obs.grandeur_hydro}")
```
